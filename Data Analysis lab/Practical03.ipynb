## Practical 03: Summarize Mobile Usage Patterns Using Descriptive Statistics

### Objective
To analyze mobile usage data and summarize patterns using descriptive statistics such as:
- Mean
- Median
- Maximum
- Minimum
- Standard Deviation

```python
import pandas as pd
```

```python
data = {
    "User": ["A", "B", "C", "D", "E"],
    "Screen_Time_Hours": [5, 7, 4, 6, 8],
    "Data_Usage_GB": [2.5, 3.2, 1.8, 2.9, 3.5],
    "Calls_Minutes": [30, 45, 20, 35, 50],
    "App_Usage_Count": [20, 25, 15, 22, 28]
}

df = pd.DataFrame(data)

df
 

```

```python
df.describe()
```

```python
mean_values = df.mean(numeric_only=True)

print("Mean Values:")
print(mean_values)
```

```python
median_values = df.median(numeric_only=True)

print("Median Values:")
print(median_values)
```

```python
max_values = df.max(numeric_only=True)

print("Maximum Values:")
print(max_values)
```

```python
min_values = df.min(numeric_only=True)

print("Minimum Values:")
print(min_values)
```

```python
std_values = df.std(numeric_only=True)

print("Standard Deviation:")
print(std_values)
```

## Result

Successfully summarized mobile usage patterns using descriptive statistics such as mean, median, minimum, maximum, and standard deviation.
