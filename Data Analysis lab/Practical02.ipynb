## Practical 02: Import and Summarize COVID-19 Cases Data

### Objective
To import COVID-19 dataset and summarize:
- Total Confirmed Cases
- Total Deaths
- Total Recovered Cases
- Recovery Rate

```python
import pandas as pd

data = {
    "Country": ["India", "USA", "Brazil", "UK", "Russia"],
    "Confirmed": [45000000, 103000000, 38000000, 24000000, 22000000],
    "Deaths": [533000, 1120000, 700000, 230000, 395000],
    "Recovered": [44400000, 100000000, 37000000, 23500000, 21400000]
}

df = pd.DataFrame(data)

df

```

```python
total_cases = df["Confirmed"].sum()

print("Total Confirmed Cases:", total_cases)
```

```python
total_deaths = df["Deaths"].sum()

print("Total Deaths:", total_deaths)
```

```python
total_recovered = df["Recovered"].sum()

print("Total Recovered:", total_recovered)
```

```python
recovery_rate = (total_recovered / total_cases) * 100

print("Recovery Rate: {:.2f}%".format(recovery_rate))
```

```python
summary = {
    "Total Cases": total_cases,
    "Total Deaths": total_deaths,
    "Total Recovered": total_recovered,
    "Recovery Rate (%)": recovery_rate
}

summary
```

## Result

Successfully imported COVID-19 dataset and calculated:
- Total confirmed cases
- Total deaths
- Total recovered cases
- Recovery rate
