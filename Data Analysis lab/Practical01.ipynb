# Data Analytics Lab
## Practical 01: Explore Student Grades Dataset

### Objective:
To calculate:
- Average Marks
- Highest Marks
- Lowest Marks
- Student-wise Total and Average


```python
import pandas as pd
data = {
    "RollNo": [1, 2, 3, 4, 5],
    "Name": ["Amit", "Neha", "Rahul", "Priya", "Karan"],
    "Maths": [78, 92, 67, 88, 55],
    "Science": [85, 79, 72, 91, 60],
    "English": [88, 95, 70, 84, 58]
}

df = pd.DataFrame(data)

df
```

```python
average_marks = df[["Maths", "Science", "English"]].mean()
average_marks
```

```python
highest_marks = df[["Maths", "Science", "English"]].max()
highest_marks
```

```python
lowest_marks = df[["Maths", "Science", "English"]].min()
lowest_marks
```

```python
df["Total"] = df[["Maths", "Science", "English"]].sum(axis=1)
df["Average"] = df["Total"] / 3

df[["RollNo", "Name", "Total", "Average"]]
```

## Result:
Successfully calculated:
- Average Marks
- Highest Marks
- Lowest Marks
- Student-wise Total and Average
