## Practical 04: Relationship Between Study Hours and Exam Scores using Linear Regression

### Objective
To analyze the relationship between study hours and exam scores using Linear Regression and predict exam scores based on study hours.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Exam_Score": [35, 40, 50, 55, 65, 70, 80, 90]
}

df = pd.DataFrame(data)

df
```

```python
X = df[["Study_Hours"]]
y = df["Exam_Score"]
```

```python
model = LinearRegression()

model.fit(X, y)
```

```python
predicted_scores = model.predict(X)

predicted_scores
```
